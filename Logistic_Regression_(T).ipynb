5 . What is Regularization in Logistic Regression ? Why is it needed ?

Regularization is a technique used to prevent overfitting in Logistic Regression (or any machine learning model).

It does this by penalizing large values of the model’s parameters (weights). This keeps the model simpler and more generalizable.

- In logistic regression, if your model tries to fit the training data too perfectly, it might memorize noise — leading to poor performance on new data.

- Regularization helps the model focus on the most important features by shrinking less useful weights toward zero.

6 . Explain the difference between Lasso, Ridge, and Elastic Net regression ?

**1. Ridge Regression (L2 Regularization)**

- Key Idea:
   - Penalizes large coefficients, but does not make them zero

  - All features are kept in the model

- When to Use:
   - When all features are useful but you want to avoid overfitting

  - Works well when features are correlated


**2. Lasso Regression (L1 Regularization)**

- Key Idea:
  - Penalizes absolute values of coefficients

  - Some coefficients can become exactly zero → useful for feature selection

- When to Use:
  - When you believe only a few features are important

  - Helps create a sparse model

**3. Elastic Net Regression (Combination of L1 and L2)**
- Key Idea:
  - Mixes the sparsity of Lasso and the stability of Ridge

  - Useful when there are many correlated features and some feature selection is desired

- When to Use:
  - When Lasso alone is too aggressive (removes too many features)

  - When Ridge keeps too many unimportant features

16 . What are the advantages and disadvantages of Logistic Regression ?

**Advantages of Logistic Regression**
1. Simple and Interpretable
   - Easy to understand and implement.

  - Coefficients show the effect of each feature on the outcome.

2. Fast to Train
  - Computationally efficient for small to medium datasets.

   - Less resource-intensive than complex models like Random Forests or Neural Networks.

3. Probabilistic Output
  - Provides class probabilities (using sigmoid or softmax).

   - Useful for risk estimation or threshold-based decisions.

4. Works Well with Linearly Separable Data
  - Performs well when the decision boundary is approximately linear.

5. Can Handle Binary and Multiclass Classification
  - Extended using One-vs-Rest (OvR) or Multinomial (Softmax) methods.

6. Regularization Support
   - Supports L1, L2, and ElasticNet regularization to prevent overfitting.

7. Good Baseline Model
  - Often used as a baseline before trying complex models.

**Disadvantages of Logistic Regression**
1. Assumes Linear Relationship
  - Assumes a linear relationship between features and the log-odds of the target.

  - Fails to capture complex, nonlinear patterns.

2. Not Suitable for Large Feature Sets Without Regularization
  - Can overfit if many features are irrelevant or correlated.

   - Needs feature selection or regularization for high-dimensional data.

3. Sensitive to Outliers
  - Logistic regression can be influenced by extreme values, especially in small datasets.

4. Poor Performance on Nonlinear or Complex Data
  - Struggles with datasets where classes are not linearly separable.

   - Complex models (like SVMs, Decision Trees, or Neural Networks) often outperform it here.

5. Requires Proper Feature Scaling
  - Some solvers (e.g., sag, lbfgs) converge faster if features are standardized.

6. Limited to Binary or Simple Multiclass Tasks
  - Not designed for multi-label classification or structured output tasks like sequences or trees.
